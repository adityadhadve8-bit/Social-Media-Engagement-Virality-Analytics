### Exploratory Data Analysis(EDA)

#### Q1. Which platform generates the highest average engagement rate?

```python
plt.figure(figsize=(14,8))
platform_eng = df.groupby("platform")["engagement_rate"].mean().sort_values(ascending= False)
sns.barplot(y= platform_eng.values, x = platform_eng.index,palette= "viridis")
plt.title('Average Engagement Rate by Platform')
plt.show()
```

##### Insight
- The platform with the highest average engagement rate that is (Instagram) demonstrates stronger audience interaction relative to content exposure.
This suggests that platform-specific content strategies play a significant role in driving user engagement.

#### Q2. Which content type gets the highest views?

```python
plt.figure(figsize=(14,8))
highest_views = df.groupby("content_type")["views"].mean().sort_values(ascending=False)
highest_views


sns.barplot(x=highest_views.values,y=highest_views.index,palette ="PuBu")
plt.title("Highest_Views By Content Type")
plt.show()
```

##### Insight
- Posts consistently outperform others in terms of visibility.
Short-form and visually engaging content typically attract larger audiences due to platform algorithms and user consumption patterns.
Content creators can leverage these formats to maximize reach and discoverability.

#### Q3. Distribution of Views

```python
plt.figure(figsize=(14,8))
sns.histplot(df["views"],bins=20,kde=True)
plt.show()
```

##### Insight
- The distribution of views is likely right-skewed, indicating that a small number of posts receive exceptionally high visibility while most posts perform at average levels.
- This pattern is common in social media ecosystems where viral content contributes disproportionately to total views.

#### Q4. Relationship Between Followers and Views

```python
plt.figure(figsize=(14,8))
sns.scatterplot(data=df, x='follower_count', y='views', alpha=0.5,hue="platform")
plt.title("Relationship Between Followers and Views")
plt.show()
```

##### Insights
- A positive relationship suggests that larger audiences contribute to greater visibility.
- However, the presence of scattered data points indicates that follower count alone does not guarantee high performance.

#### Q5. Impact of Trending Audio on Engagement

```python
plt.figure(figsize=(14,8))
sns.boxplot(data=df,x='trending_audio', y='engagement_rate',palette="PuBu")
plt.title("Impact of Trending Audio on Engagement")
plt.show()
```

##### Insight
- Posts utilizing trending audio generally exhibit higher engagement levels compared to those without trending audio.
- This suggests that platform trends can enhance content discoverability and audience interaction.
- Leveraging trending elements may increase the likelihood of content amplification.

#### Q6. Does CTA Increase Engagement?

```python
plt.figure(figsize=(14,8))
sns.boxplot(data=df,
            x='call_to_action',
            y='engagement_rate')
plt.show()
```

##### Insight
- Posts containing a Call-To-Action tend to encourage more user participation.
- Explicit prompts such as liking, commenting, sharing, or subscribing can positively influence engagement metrics.

#### Q7. Which Emotion Trigger Performs Best?

```python
plt.figure(figsize=(14,8))
plt.title("Best Emotion Trigger")
emotions = df.groupby("emotion_trigger")["engagement_rate"].mean().sort_values(ascending=False)

emotions.plot(kind="bar",color="green")
plt.show()
```

##### Insight
- Content designed around emotions such as Fear , Funny and neutral often generates higher engagement.
- Understanding audience psychology can help optimize content strategies.

#### Q8. Best Posting Hour

```python
plt.figure(figsize=(14,8))
plt.title("Best Posting Hour")
hourly = df.groupby('post_time')['engagement_rate'].mean()

hourly.plot(marker='o')
plt.show()
```

##### Insight
- 6:00PM is considered to be the best posting hour. 
- Identifying peak engagement hours allows creators to schedule content for maximum visibility.
- Posting during high-activity periods may improve both reach and engagement.

#### Q9. Best Day of Week for Engagement

```python
plt.figure(figsize=(14,8))
plt.title("Best Day of Week for Engagement")
day = df.groupby('day_of_week')['engagement_rate'].mean()

day.plot(marker='x',color= "orange")
plt.show()
```

##### Insight
- User interaction fluctuates across different days of the week.
- Certain days consistently outperform others like Friday and Saturday indicating favorable audience availability.
- Scheduling content around these periods can improve performance outcomes.

#### Q10. Correlation Between Numerical Variables

```python
plt.figure(figsize=(14,8))

sns.heatmap(
    df.select_dtypes(include=np.number).corr(),
    cmap='Blues',
    annot=False
)

plt.show()
```

##### Insight
- Strong positive correlations between retention, watch time, engagement, and virality indicate interconnected performance drivers.

#### Q11. Viral vs Non-Viral Posts

```python
df['viral'].value_counts().plot(kind='pie',autopct='%1.1f%%',legend=  True)
plt.title("Viral vs Non-Viral Posts")
plt.ylabel('')
plt.show()
```

##### Insight
- Viral content typically represents a smaller proportion of total posts.
- This highlights the competitive nature of social media and the difficulty of achieving viral reach.

#### Q12. Platform Share of Total Posts

```python
df['platform'].value_counts().plot( kind='pie', autopct='%1.1f%%')
plt.title("Total Post Across Different Platform")
plt.ylabel('')
plt.show()
```

```python
# Using Plotly
import plotly.express as px

platform_dist = (
    df['platform']
    .value_counts()
    .reset_index()
)

platform_dist.columns = ['platform', 'count']

fig = px.pie(
    platform_dist,
    names='platform',
    values='count',
    title='Distribution of Posts Across Social Media Platforms',
    hole=0.3
)

fig.update_traces(
    textinfo='percent+label'
)

fig.show()
```

##### Insights
- YouTube has the highest number of post distribution across all social media platforms followed by Tiktok and Instagram

#### Q13. Which content category receives the highest average engagement rate across all social media platforms?

```python
import plotly.express as px

category_eng = (
    df.groupby('category')['engagement_rate']
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

fig = px.bar(
    category_eng,
    x='category',
    y='engagement_rate',
    title='Average Engagement Rate by Content Category',
    text_auto='.2%',
    color='engagement_rate'
)

fig.update_layout(
    xaxis_title='Content Category',
    yaxis_title='Average Engagement Rate',
    title_x=0.5
)

fig.show()
```

##### Insight
- Motivational and Education category consistently attract more audience interaction than others.
- High-performing categories reflect audience interests and platform trends.
- Focusing on these categories can improve engagement and content effectiveness.

#### Q14. How do engagement, audience retention, content reach, virality, and content characteristics vary across the dataset?

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(3, 3, figsize=(20, 15))

# 1. Bar Plot
platform_eng = df.groupby('platform')['engagement_rate'].mean()
platform_eng.plot(kind='bar', ax=axes[0,0])
axes[0,0].set_title('Average Engagement Rate by Platform')
axes[0,0].set_ylabel('Engagement Rate')

# 2. Histogram
sns.histplot(df['views'], bins=30, kde=True, ax=axes[0,1])
axes[0,1].set_title('Distribution of Views')

# 3. Scatter Plot
sns.scatterplot(
    data=df,
    x='follower_count',
    y='views',
    alpha=0.6,
    ax=axes[0,2]
)
axes[0,2].set_title('Followers vs Views')

# 4. Box Plot
sns.boxplot(
    data=df,
    x='trending_audio',
    y='engagement_rate',
    ax=axes[1,0]
)
axes[1,0].set_title('Trending Audio vs Engagement')

# 5. Line Chart
hourly = df.groupby('post_time')['engagement_rate'].mean()
hourly.plot(marker='o', ax=axes[1,1])
axes[1,1].set_title('Best Posting Hour')

# 6. Pie Chart
df['viral'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    ax=axes[1,2]
)
axes[1,2].set_title('Viral vs Non-Viral Posts')
axes[1,2].set_ylabel('')

# 7. Violin Plot
sns.violinplot(
    data=df,
    x='viral',
    y='completion_rate',
    ax=axes[2,0]
)
axes[2,0].set_title('Completion Rate by Virality')

# 8. Count Plot
sns.countplot(
    data=df,
    x='content_type',
    ax=axes[2,1]
)
axes[2,1].set_title('Distribution of Content Types')
axes[2,1].tick_params(axis='x', rotation=45)

# 9. KDE Plot
sns.kdeplot(
    data=df,
    x='watch_time_avg',
    fill=True,
    ax=axes[2,2]
)
axes[2,2].set_title('Watch Time Density')

plt.suptitle(
    'Social Media Engagement & Virality Analytics Dashboard',
    fontsize=20,
    fontweight='bold'
)

plt.tight_layout()
plt.show()
```

- The dashboard provides a holistic view of social media performance by combining engagement, retention, reach, virality, and content distribution metrics.
- Multiple visualizations reveal that successful content is typically characterized by strong audience retention, high engagement, and effective content strategy.
- Audience behavior, content format, emotional appeal, and posting strategy collectively contribute to content success.
